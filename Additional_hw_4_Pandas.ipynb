6. Трансформуйте колонки 'Ваш вік', 'Зарплата у $$$ за місяць, лише ставка після сплати податків' у тип даних float. УВАГА! Перед трансформацією подивіться, які дані містять ці колонки. Аби успішно трансформувати тип даних, треба спочатку привести рядки до того виду, який буде підходити, аби потім рядок зробити типом float.

```python
for col in ['Ваш вік', 'Зарплата у $$$ за місяць, лише ставка після сплати податків']:
    df[col] = df[col].replace({r'\D': ''}, regex=True)
```

```python
for col in ['Ваш вік', 'Зарплата у $$$ за місяць, лише ставка після сплати податків']:
    df[col] = pd.to_numeric(df[col]).astype(float)
```

```python
df.Timestamp
```

```python
def correct_date_time(date):
    if len(date) == 19:
        return datetime.datetime.strptime(date, '%m.%d.%Y %H:%M:%S')
    elif len(date) == 16:
        return datetime.datetime.strptime(date, '%d.%m.%Y %H:%M')
    else:
        return None
```

```python
df['Timestamp'].apply(correct_date_time)
```

11. Є така штука як сортування значень категоріальної змінної. Нам це зараз стане в нагоді, адже рівні англійської можна посортувати.
Давайте відсотруємо рівні англ. і тайтли. Просто запустіть код нижче і проаналізуйте, що він робить.

```python
df['Знання англійської мови'] = pd.Categorical(df['Знання англійської мови'],
                      categories=['Elementary','Pre-Intermediate','Intermediate','Upper-Intermediate','Advanced'],
                      ordered=True)

df['Ваш тайтл'] = pd.Categorical(df['Ваш тайтл'],
                      categories=['Intern/Trainee', 'Junior', 'Немає тайтлу', 'Middle', 'Senior', 'Tech Lead', 'Team Lead', 'Architect', 'Head', 'Manager'],
                      ordered=True)
```

**Цей код сортує рівні англійської і тайтли відповідно до загально прийнятої логіки: від найнижчого до найбільшого.**

12. Зробіть зводну таблицю, де стовпці - це 'Ваш тайтл', рядки - 'Знання англійської мови', а значення - медіанна ЗП. Які висновки можна зробити?
Тут ще для зручності можна зробити градієнтне забарвлення і відформатувати значення для легшого перегляду.

```python
median_salary_by_title_eng = df.pivot_table(index = 'Знання англійської мови', columns = 'Ваш тайтл', values = 'Зарплата у $$$ за місяць, лише ставка після сплати податків', aggfunc = {'Зарплата у $$$ за місяць, лише ставка після сплати податків': ['median']}, observed=False)
display(median_salary_by_title_eng.style.format("{:.0f}").background_gradient(low=0.9, high=1.0))
```

**Висновки:** 
- цікаво, що інтерни як і з рівнем англійської Elementary, так і Advanced мають однакову медіану ЗП;
- для інших тайтлів досить закономірно, що працівники з вищим рівнем англійської мають більшу медіану ЗП;
- проте для Tech Lead люди з рівнем англійської Elementary мають таку ж ЗП, як і люди з вищим рівнем англійської. У цьому випадку заробітна плата мабуть більше залежить від технічних навичок і рівня відповідальності.

13. Виведіть результат методу describe() по колонцы розміру ЗП для топ 5 за частотою зустрічання в даних областей (В якій області ви зараз живете?). Зробіть висновки, де вигідніше жити з точки зору ЗП і чи суттєва різниця на Ваш погляд для прийняття рішення про переїзд?

```python
top_5_region = df['В якій області ви зараз живете?'].value_counts().head(5).index
```

```python
for region in top_5_region:
    print(f"Cтатистика для регіону: {region}")
    region_data = df[df['В якій області ви зараз живете?'] == region]
    print(region_data['Зарплата у $$$ за місяць, лише ставка після сплати податків'].describe())
    print("\n")
```

## **Висновки:** 
- найбільші значення середньої заробітної плати ми маємо у Київській та Вінницькій областях (водночас для Вінницької ми маємо набагато менше значень);
-  у Вінницькій області ми маємо найбільше максимальне значення ЗП серед усіх 5 областей;
- найбільша мінімальна ЗП у Львівській області, тому мабуть розпочинати свою кар'єру найвигідніше саме там;
- якщо обирати серед міст-мільйоників, то найкращі ЗП у Києві або Львові, серед менших міст - варто обирати Вінницю.

14. Як зростає зарплата з роками? Для цього нам треба відформатувати колонку "Загальний стаж роботи за спеціальністю" аналогічно до того, як ми це робили для колонки 'Знання англійської мови' і потім вивести медіанні значення для кількості років стажу.

```python
df['Загальний стаж роботи за спеціальністю'].unique()
```

```python
df['Загальний стаж роботи за спеціальністю'] = pd.Categorical(df['Загальний стаж роботи за спеціальністю'],
    categories=[
        'Менше як 3 місяці', '3 місяці', 'Пів року', '1 рік', '1,5 року', 
        '2 роки', '3 роки', '4 роки', '5 років', '6 років', '7 років', 
        '8 років', '9 років', '10 років', '11 років', '12 років', 
        '13 років', '14 років', '15 і більше років'
    ],
    ordered=True
)
```

```python
median_salary_by_years = df.groupby('Загальний стаж роботи за спеціальністю', observed=False)['Зарплата у $$$ за місяць, лише ставка після сплати податків'].median()
display(median_salary_by_years)
```

**Висновки:** 
- цікаво, що у опитуваних працівників, які працюють менше як 3 місяці медіана ЗП більша ніж у тих, хто працює 3 місяці і навіть пів року;
- далі, починаючи з пів року і до 8 років, медіана зарплата постійно зростає разом із стажем;
- з 9 і до понад 15 років ЗП змінюється якось нелінійно (тобто і з 11, і з 14 роками стажу можна мати ЗП близько 5000).
