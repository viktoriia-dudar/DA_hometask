````python
def calculate_variance(numbers_list):
    sum_list = 0
    for number in numbers_list:
        sum_list += number  # Сума елементів

    mean = sum_list / len(numbers_list)  # Середнє значення

    sum_squares = 0
    for number in numbers_list:
        sum_squares += number ** 2  # Сума квадратів елементів

    mean_squares = sum_squares / len(numbers_list)  # Середнє значення квадратів

    variance = mean_squares - mean ** 2  # Дисперсія

    return variance
````
